```python
class test():
    """ The test class will return all the binary operations described for two rational numbers and we can see 
    for ourselves if our function rationaliser works. This is my adopted test to see if my class works  """
    
    def  __init__( self, numb1,numb2,precision=1.e-5):
        
        self.add,self.sub,self.eq,self.gt,self.lt,self.mul,self.div = self.work(numb1,numb2)
    def work(self,numb1,numb2):
        """I define my function "rationaliser" within "work " and use it to obtain the binary operations """
         
        def rationaliser(numb, precision=1.e-5 ):
            """Documentation explained in Exercise_4.py before"""
            if abs(numb - int(numb))<precision:
                sign = 1
                N= int(numb)
                D=1
                flag = 4
                return(round(sign*N),round(D))
    
            else:
                x=abs(numb)
                sign = x/numb
                i=0
                X=[]
                flag=0
                for i in range(10): 
                    x_int = int(x) # factoring the integer values of x
                    diff = abs(x_int - x)
                    if diff < precision: 
                        X.append(x_int)# Voila we have already reached the number somehow and also completed our list 
                        flag =3
                        if len(X)<3:
                            Y = X[::-1]
                            N =  Y[0]*Y[1]+1
                            D = Y[0]
                            flag =3
                            break
                    else: 
                        y = 1/diff
                        X.append(x_int)
                        x=y # added new here
                    if len(X)<2:
                        continue
                    Y = X[::-1] # This reverses the list
                    Z=[]
                    j=0
                    for j in range(len(X)):
                        if j== (len(X)-1):
                            break
                        if j == 0:
                            z = Y[j]*Y[j+1] + 1
                            Z.append(z)
                        else:
                            z = Y[j]*Y[j+1] + Y[j-1]
                            Z.append(z)
                        Y[j+1] = Z[j]
                    N = Z[j-1]
                    D = Z[j-2]
                    frac = N/D
                    prec = abs(frac - abs(numb))
                    if prec < precision:
                        flag=1
                        break
                if flag==1 or 2 or 3 or 4:          
                    return(reducer(round(sign*N),round(D)))

    
            
        n1,d1 = list(rationaliser(numb1))
        n2,d2 = list(rationaliser(numb2))
        
        """Here are my binary operations """
        
        a = f'Addition: {list(rationaliser(n1/d1 + n2/d2))[0]}/{list(rationaliser(n1/d1 + n2/d2))[1]}'
        s= f'Subtraction: {list(rationaliser(numb1-numb2))[0]}/{list(rationaliser(numb1-numb2))[1]}'
        e = f'Equality?: {(n1,d1) == (n2,d2)}'
        g = f'1st greater than 2nd?: {n1/d1 > n2/d2}'
        l = f'1st lesser than 2nd?: {n1/d1 < n2/d2}'
        m = f'Multiplication: {list(rationaliser(n1*n2/(d1*d2)))[0]}/{list(rationaliser(n1*n2/(d1*d2)))[1]}'
        if n2 !=0:
            d = f'Division: {list(rationaliser(numb1/numb2))[0]}/{list(rationaliser(numb1/numb2))[1]}'
        else:
            d = 'Error Division by 0'
        return(a,s,e,g,l,m,d)
                
    def __repr__ (self) :
        return f'{type(self).__name__}{self.add,}\n {self.sub,}\n{self.eq,}\n{self.gt,}\n{self.lt,}\n{self.mul,}\n{self.div}'
    def __str__(self):
        return f'{self.add}\n{self.sub}\n{self.eq}\n{self.gt}\n{self.lt}\n{self.mul}\n{self.div} \n'
        
        
        
```

```python
p= test(-1.2,-1.3) # Check how negative numbers work 
print(p)
q = test(2,3) # Check how integers work
print(q)
r= test (1.2,1.200001) # Check whether approximate numbers are equal (as rational numbers)
print(r)
s= test(1,0) # Check how 0 works 
print(s)
t = test(1.6755,10.6576) # Check how random numbers work 
print(t)
```
